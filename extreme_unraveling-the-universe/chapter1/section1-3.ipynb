```python
#!/usr/bin/env python3
"""
Hydrogen Density Analysis - Solution Check
水素密度解析 - 解答チェック

Problem: 水素原子（陽子＋電子）の質量を mH とすると，中性水素ガスの質量密度 ρ と個数密度 N の関係はどうなるか？　また完全電離したプラズマではどうなるか？
Given Solution: ρ = mHN，ρ = mHN/2

This script analyzes whether the given solution is correct based on physics principles.
"""

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib import rcParams
import seaborn as sns

# Set style for better plots
plt.style.use('seaborn-v0_8')
rcParams['figure.figsize'] = (12, 8)
rcParams['font.size'] = 10

# Physical constants
m_proton = 1.67262192e-27  # kg (proton mass)
m_electron = 9.1093837e-31  # kg (electron mass)
m_H = m_proton + m_electron  # Total hydrogen atom mass

def print_problem_statement():
    """Print the problem statement and given solution"""
    print("=" * 60)
    print("HYDROGEN DENSITY ANALYSIS - SOLUTION CHECK")
    print("水素密度解析 - 解答チェック")
    print("=" * 60)
    print()
    
    print("PROBLEM:")
    print("水素原子（陽子＋電子）の質量を mH とすると，")
    print("中性水素ガスの質量密度 ρ と個数密度 N の関係はどうなるか？")
    print("また完全電離したプラズマではどうなるか？")
    print()
    
    print("TRANSLATION:")
    print("If the mass of a hydrogen atom (proton + electron) is mH,")
    print("what is the relationship between mass density ρ and number density N")
    print("of neutral hydrogen gas? Also, what happens in completely ionized plasma?")
    print()
    
    print("GIVEN SOLUTION:")
    print("ρ = mHN，ρ = mHN/2")
    print()
    
    print("PHYSICAL CONSTANTS:")
    print(f"Proton mass: {m_proton:.3e} kg")
    print(f"Electron mass: {m_electron:.3e} kg")
    print(f"Hydrogen atom mass (mH): {m_H:.3e} kg")
    print(f"Mass ratio (proton/electron): {m_proton/m_electron:.0f}")
    print()

def analyze_mass_conservation():
    """Analyze mass conservation during ionization"""
    print("MASS CONSERVATION ANALYSIS")
    print("質量保存解析")
    print("-" * 40)
    
    # Before ionization
    print("BEFORE IONIZATION (中性水素):")
    print("• 1 hydrogen atom = 1 proton + 1 electron (bound)")
    print("• Mass per atom = m_proton + m_electron = mH")
    print("• For N atoms: Total mass = N × mH")
    print()
    
    # After ionization
    print("AFTER IONIZATION (電離プラズマ):")
    print("• 1 hydrogen atom becomes: 1 proton + 1 electron (separated)")
    print("• Mass = m_proton + m_electron = mH (SAME!)")
    print("• For N original atoms: Total mass = N × mH (SAME!)")
    print()
    
    # Conservation law
    print("CONSERVATION OF MASS LAW:")
    print("質量保存の法則")
    print("• Ionization only SEPARATES particles - it doesn't DESTROY mass!")
    print("• 電離は粒子を分離するだけ - 質量は破壊されない！")
    print()
    
    return True

def calculate_densities():
    """Calculate densities using both correct physics and given solution"""
    # Create range of number densities
    N_range = np.logspace(15, 20, 50)  # atoms/m³
    
    # Correct physics: Both cases have ρ = N × mH
    rho_neutral_correct = N_range * m_H
    rho_ionized_correct = N_range * m_H
    
    # Given solution
    rho_neutral_given = N_range * m_H      # ρ = mHN
    rho_ionized_given = N_range * m_H / 2  # ρ = mHN/2
    
    return N_range, rho_neutral_correct, rho_ionized_correct, rho_neutral_given, rho_ionized_given

def plot_density_comparison():
    """Plot density comparison between correct physics and given solution"""
    N_range, rho_neutral_correct, rho_ionized_correct, rho_neutral_given, rho_ionized_given = calculate_densities()
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))
    
    # Plot 1: Correct Physics
    ax1.loglog(N_range, rho_neutral_correct, 'b-', linewidth=3, label='Neutral H (ρ = N×mH)')
    ax1.loglog(N_range, rho_ionized_correct, 'b--', linewidth=3, label='Ionized H (ρ = N×mH)')
    ax1.set_xlabel('Number Density N (atoms/m³)')
    ax1.set_ylabel('Mass Density ρ (kg/m³)')
    ax1.set_title('CORRECT PHYSICS\n(Mass Conservation)')
    ax1.grid(True, alpha=0.3)
    ax1.legend()
    
    # Add annotation
    ax1.annotate('Both lines overlap!\nρ = N × mH', 
                xy=(1e18, 1e-9), xytext=(1e16, 1e-7),
                arrowprops=dict(arrowstyle='->', color='green', lw=2),
                fontsize=12, ha='center', color='green', fontweight='bold',
                bbox=dict(boxstyle="round,pad=0.3", facecolor="lightgreen", alpha=0.7))
    
    # Plot 2: Given Solution
    ax2.loglog(N_range, rho_neutral_given, 'r-', linewidth=3, label='Neutral H (ρ = N×mH)')
    ax2.loglog(N_range, rho_ionized_given, 'r--', linewidth=3, label='Ionized H (ρ = N×mH/2)')
    ax2.set_xlabel('Number Density N (atoms/m³)')
    ax2.set_ylabel('Mass Density ρ (kg/m³)')
    ax2.set_title('GIVEN SOLUTION\n(Mass Loss?)')
    ax2.grid(True, alpha=0.3)
    ax2.legend()
    
    # Add annotation
    ax2.annotate('Mass reduces by half?\nViolates conservation!', 
                xy=(1e18, 5e-10), xytext=(1e16, 1e-8),
                arrowprops=dict(arrowstyle='->', color='red', lw=2),
                fontsize=12, ha='center', color='red', fontweight='bold',
                bbox=dict(boxstyle="round,pad=0.3", facecolor="lightcoral", alpha=0.7))
    
    plt.tight_layout()
    plt.show()
    
    return fig

def create_particle_diagram():
    """Create a visual diagram showing particle composition"""
    fig, ax = plt.subplots(figsize=(14, 8))
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 8)
    ax.set_aspect('equal')
    
    # Title
    ax.text(5, 7.5, 'Particle Composition per Original Hydrogen Atom', 
            ha='center', va='center', fontsize=16, fontweight='bold')
    
    # Neutral Hydrogen side
    ax.text(2.5, 6.5, 'NEUTRAL HYDROGEN\n中性水素', ha='center', va='center', 
            fontsize=14, fontweight='bold', color='blue')
    
    # Draw hydrogen atoms
    for i in range(3):
        for j in range(2):
            x, y = 1.5 + i*0.8, 5.5 - j*0.8
            # Hydrogen atom
            circle = patches.Circle((x, y), 0.25, facecolor='lightblue', edgecolor='blue', linewidth=2)
            ax.add_patch(circle)
            ax.text(x, y, 'H', ha='center', va='center', fontweight='bold', fontsize=10)
    
    # Neutral hydrogen info
    ax.text(2.5, 3.5, '• 1 Hydrogen atom\n• 1 Proton + 1 Electron (bound)\n• Mass = mH\n• Total particles = 1', 
            ha='center', va='top', fontsize=11, 
            bbox=dict(boxstyle="round,pad=0.5", facecolor="lightblue", alpha=0.3))
    
    # Formula
    ax.text(2.5, 2.5, 'ρ = N × mH', ha='center', va='center', fontsize=14, fontweight='bold',
            bbox=dict(boxstyle="round,pad=0.3", facecolor="white", edgecolor="blue", linewidth=2))
    
    # Ionized Plasma side
    ax.text(7.5, 6.5, 'IONIZED PLASMA\n電離プラズマ', ha='center', va='center', 
            fontsize=14, fontweight='bold', color='red')
    
    # Draw separated protons and electrons
    for i in range(3):
        for j in range(2):
            x, y = 6.5 + i*0.8, 5.5 - j*0.8
            # Proton
            circle_p = patches.Circle((x-0.15, y), 0.15, facecolor='pink', edgecolor='red', linewidth=2)
            ax.add_patch(circle_p)
            ax.text(x-0.15, y, 'p+', ha='center', va='center', fontweight='bold', fontsize=8)
            
            # Electron
            circle_e = patches.Circle((x+0.15, y), 0.1, facecolor='yellow', edgecolor='orange', linewidth=2)
            ax.add_patch(circle_e)
            ax.text(x+0.15, y, 'e-', ha='center', va='center', fontweight='bold', fontsize=8)
    
    # Ionized plasma info
    ax.text(7.5, 3.5, '• 1 Proton + 1 Electron (separated)\n• Mass = mH (SAME!)\n• Total particles = 2\n• Total mass = SAME!', 
            ha='center', va='top', fontsize=11,
            bbox=dict(boxstyle="round,pad=0.5", facecolor="lightcoral", alpha=0.3))
    
    # Formula
    ax.text(7.5, 2.5, 'ρ = N × mH', ha='center', va='center', fontsize=14, fontweight='bold',
            bbox=dict(boxstyle="round,pad=0.3", facecolor="white", edgecolor="red", linewidth=2))
    
    # Conservation arrow
    ax.annotate('', xy=(6.2, 4), xytext=(3.8, 4),
                arrowprops=dict(arrowstyle='<->', color='green', lw=3))
    ax.text(5, 4.3, 'MASS CONSERVED\n質量保存', ha='center', va='center', 
            fontsize=12, fontweight='bold', color='green')
    
    # Same mass emphasis
    ax.text(5, 1.5, 'SAME TOTAL MASS = N × mH', ha='center', va='center', 
            fontsize=16, fontweight='bold', color='green',
            bbox=dict(boxstyle="round,pad=0.5", facecolor="lightgreen", alpha=0.7))
    
    ax.set_xticks([])
    ax.set_yticks([])
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    
    plt.tight_layout()
    plt.show()
    
    return fig

def analyze_given_solution():
    """Analyze the problems with the given solution"""
    print("ANALYSIS OF GIVEN SOLUTION")
    print("与えられた解答の分析")
    print("-" * 40)
    
    print("GIVEN SOLUTION STATES:")
    print("• Neutral hydrogen: ρ = mHN ✅ (correct)")
    print("• Ionized plasma: ρ = mHN/2 ❌ (incorrect)")
    print()
    
    print("PROBLEMS WITH ρ = mHN/2:")
    print("1. ❌ VIOLATES CONSERVATION OF MASS")
    print("   - Suggests ionization reduces mass by half")
    print("   - Mass cannot be destroyed during ionization")
    print()
    
    print("2. ❌ CONTRADICTS PHYSICAL REALITY")
    print("   - Before: 1 proton + 1 electron = mH")
    print("   - After: 1 proton + 1 electron = mH (same mass!)")
    print()
    
    print("3. ❌ UNPHYSICAL RESULT")
    print("   - Where did the 'missing' mass go?")
    print("   - Energy-mass equivalence doesn't apply here")
    print()
    
    # Calculate the implied mass loss
    N_example = 1e18  # atoms/m³
    rho_correct = N_example * m_H
    rho_given = N_example * m_H / 2
    mass_loss_fraction = (rho_correct - rho_given) / rho_correct
    
    print(f"QUANTITATIVE ANALYSIS:")
    print(f"• For N = {N_example:.1e} atoms/m³:")
    print(f"  - Correct: ρ = {rho_correct:.2e} kg/m³")
    print(f"  - Given solution: ρ = {rho_given:.2e} kg/m³")
    print(f"  - Implied mass loss: {mass_loss_fraction:.0%}")
    print()

def possible_explanations():
    """Discuss possible explanations for the given solution"""
    print("POSSIBLE EXPLANATIONS FOR GIVEN SOLUTION")
    print("与えられた解答の可能な説明")
    print("-" * 40)
    
    print("1. TYPO/MISTAKE")
    print("   - The '/2' might be a printing or transcription error")
    print("   - 誤植または転記ミスの可能性")
    print()
    
    print("2. DIFFERENT DEFINITION OF N")
    print("   - If N = total number of particles (not original atoms)")
    print("   - In plasma: N_total = 2N_atoms (protons + electrons)")
    print("   - Then ρ = N_total × mH/2 = 2N_atoms × mH/2 = N_atoms × mH")
    print("   - But this is unconventional notation")
    print()
    
    print("3. CONFUSION WITH PARTICLE DENSITY")
    print("   - Mixing up mass density with particle density")
    print("   - Number density of particles doubles in plasma")
    print("   - But mass density stays the same")
    print()
    
    print("4. INCORRECT PHYSICS UNDERSTANDING")
    print("   - Misunderstanding of what happens during ionization")
    print("   - Thinking ionization affects total mass")
    print()

def numerical_examples():
    """Show numerical examples with realistic values"""
    print("NUMERICAL EXAMPLES")
    print("数値例")
    print("-" * 40)
    
    examples = [
        ("Interstellar medium", "星間物質", 1e6),
        ("Molecular cloud", "分子雲", 1e8),
        ("Stellar atmosphere", "恒星大気", 1e17),
    ]
    
    for name_en, name_jp, N in examples:
        print(f"{name_en} ({name_jp}):")
        print(f"  N = {N:.1e} atoms/m³")
        
        # Correct physics
        rho_correct = N * m_H
        print(f"  Correct: ρ = {rho_correct:.2e} kg/m³ (both neutral & ionized)")
        
        # Given solution
        rho_neutral_given = N * m_H
        rho_ionized_given = N * m_H / 2
        print(f"  Given solution:")
        print(f"    Neutral: ρ = {rho_neutral_given:.2e} kg/m³")
        print(f"    Ionized: ρ = {rho_ionized_given:.2e} kg/m³")
        print(f"    Implied mass loss: {((rho_correct - rho_ionized_given)/rho_correct)*100:.0f}%")
        print()

def final_conclusion():
    """Present the final conclusion"""
    print("FINAL CONCLUSION")
    print("最終結論")
    print("=" * 40)
    
    print("✅ CORRECT ANSWER:")
    print("• Neutral hydrogen: ρ = N × mH")
    print("• Ionized plasma: ρ = N × mH (SAME!)")
    print()
    
    print("❌ GIVEN SOLUTION ERROR:")
    print("• The solution 'ρ = mHN/2' for ionized plasma is INCORRECT")
    print("• It violates the fundamental principle of mass conservation")
    print()
    
    print("🔬 PHYSICS PRINCIPLE:")
    print("• Ionization only separates particles - it doesn't destroy mass")
    print("• Total mass density remains constant: ρ = N × mH")
    print("• Only the number of particles changes (1 → 2 particles per atom)")
    print()
    
    print("📋 RECOMMENDATION:")
    print("• Use ρ = N × mH for both neutral and ionized hydrogen")
    print("• Question solutions that violate conservation laws")
    print("• Always check if results make physical sense")
    print()

def main():
    """Main function to run the complete analysis"""
    print_problem_statement()
    
    analyze_mass_conservation()
    
    print("CREATING VISUALIZATIONS...")
    print("可視化を作成中...")
    
    # Create plots
    fig1 = plot_density_comparison()
    fig2 = create_particle_diagram()
    
    analyze_given_solution()
    
    possible_explanations()
    
    numerical_examples()
    
    final_conclusion()
    
    print("Analysis complete!")
    print("解析完了！")

if __name__ == "__main__":
    main()

# Additional utility functions for interactive use
def check_solution(rho_neutral_formula, rho_ionized_formula, N_test=1e18):
    """
    Check if given density formulas are physically correct
    
    Parameters:
    rho_neutral_formula: function that takes N and returns rho for neutral gas
    rho_ionized_formula: function that takes N and returns rho for ionized plasma
    N_test: test number density value
    """
    print(f"TESTING WITH N = {N_test:.1e} atoms/m³")
    
    rho_neutral = rho_neutral_formula(N_test)
    rho_ionized = rho_ionized_formula(N_test)
    
    print(f"Neutral: ρ = {rho_neutral:.2e} kg/m³")
    print(f"Ionized: ρ = {rho_ionized:.2e} kg/m³")
    
    if rho_neutral == rho_ionized:
        print("✅ CORRECT: Mass is conserved")
    else:
        mass_change = (rho_ionized - rho_neutral) / rho_neutral * 100
        print(f"❌ INCORRECT: Mass changes by {mass_change:.1f}%")
        print("This violates conservation of mass!")
    
    return rho_neutral == rho_ionized

# Example usage:
if __name__ == "__main__":
    # Test the given solution
    print("\n" + "="*60)
    print("TESTING GIVEN SOLUTION")
    print("="*60)
    
    # Define the formulas from given solution
    neutral_formula = lambda N: N * m_H      # ρ = mHN
    ionized_formula = lambda N: N * m_H / 2  # ρ = mHN/2
    
    is_correct = check_solution(neutral_formula, ionized_formula)
    
    if not is_correct:
        print("\nTESTING CORRECT SOLUTION:")
        correct_neutral = lambda N: N * m_H
        correct_ionized = lambda N: N * m_H
        check_solution(correct_neutral, correct_ionized)
```
