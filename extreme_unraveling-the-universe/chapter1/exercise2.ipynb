### 基本次元 / Basic Dimensions
- **長さ (Length)**: [L] → 単位: m, cm
- **質量 (Mass)**: [M] → 単位: kg, g  
- **時間 (Time)**: [T] → 単位: s, year

### 基本定数の次元 / Dimensions of Fundamental Constants
- **光速度**: [c] = LT⁻¹

## 2. プランク時間の導出 / Derivation of Planck Time

### ステップバイステップの導出 / Step-by-Step Derivation

時間の次元 [T] を持つ量を作るために、c, G, h の組み合わせを考えます。

$$G^a \cdot h^b \cdot c^d = [T]$$

次元を代入すると：
$$[L^3 M^{-1} T^{-2}]^a \cdot [L^2 M T^{-1}]^b \cdot [L T^{-1}]^d = [T]$$

```python
def derive_planck_time():
    """Show step-by-step derivation of Planck time"""
    print("プランク時間の導出 / Planck Time Derivation")
    print("=" * 50)
    print()
    
    print("ステップ1: 次元方程式の設定 / Step 1: Setting up dimensional equation")
    print("G^a × h^b × c^d = [T]")
    print()
    
    print("ステップ2: 次元を代入 / Step 2: Substituting dimensions")
    print("[L³M⁻¹T⁻²]^a × [L²MT⁻¹]^b × [LT⁻¹]^d = [T]")
    print()
    
    print("ステップ3: 指数を展開 / Step 3: Expanding exponents")
    print("L^(3a+2b+d) × M^(-a+b) × T^(-2a-b-d) = L⁰M⁰T¹")
    print()
    
    print("ステップ4: 連立方程式を解く / Step 4: Solving simultaneous equations")
    print("長さの次元 / Length: 3a + 2b + d = 0")
    print("質量の次元 / Mass: -a + b = 0")
    print("時間の次元 / Time: -2a - b - d = 1")
    print()
    
    print("ステップ5: 解 / Step 5: Solution")
    print("a = 1/2, b = 1/2, d = -5/2")
    print()
    
    print("ステップ6: 最終結果 / Step 6: Final result")
    print("tP = √(Gh/c⁵)")
    print()
    
    # Calculate numerical value
    t_planck = np.sqrt(G * h / c**5)
    print(f"数値計算 / Numerical calculation:")
    print(f"tP = {t_planck:.2e} s")
    print(f"tP ≈ {t_planck/1e-43:.1f} × 10⁻⁴³ s")
    print()
    
    return t_planck

t_planck = derive_planck_time()
```

## 3. プランク長さの導出 / Derivation of Planck Length

同様の方法でプランク長さを導出します。

```python
def derive_planck_length():
    """Show step-by-step derivation of Planck length"""
    print("プランク長さの導出 / Planck Length Derivation")
    print("=" * 50)
    print()
    
    print("ステップ1: 次元方程式の設定 / Step 1: Setting up dimensional equation")
    print("G^a × h^b × c^d = [L]")
    print()
    
    print("ステップ2: 次元を代入 / Step 2: Substituting dimensions")
    print("[L³M⁻¹T⁻²]^a × [L²MT⁻¹]^b × [LT⁻¹]^d = [L]")
    print()
    
    print("ステップ3: 指数を展開 / Step 3: Expanding exponents")
    print("L^(3a+2b+d) × M^(-a+b) × T^(-2a-b-d) = L¹M⁰T⁰")
    print()
    
    print("ステップ4: 連立方程式を解く / Step 4: Solving simultaneous equations")
    print("長さの次元 / Length: 3a + 2b + d = 1")
    print("質量の次元 / Mass: -a + b = 0")
    print("時間の次元 / Time: -2a - b - d = 0")
    print()
    
    print("ステップ5: 解 / Step 5: Solution")
    print("a = 1/2, b = 1/2, d = -3/2")
    print()
    
    print("ステップ6: 最終結果 / Step 6: Final result")
    print("ℓP = √(Gh/c³)")
    print()
    
    # Calculate numerical value
    l_planck = np.sqrt(G * h / c**3)
    l_planck_cm = l_planck * 100  # Convert to cm
    print(f"数値計算 / Numerical calculation:")
    print(f"ℓP = {l_planck:.2e} m")
    print(f"ℓP = {l_planck_cm:.2e} cm")
    print(f"ℓP ≈ {l_planck_cm/4e-33:.1f} × 4 × 10⁻³³ cm")
    print()
    
    return l_planck

l_planck = derive_planck_length()
```
