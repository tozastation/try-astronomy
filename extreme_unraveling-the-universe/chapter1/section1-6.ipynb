```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Wedge, Arc
import math

# Constants
deg_to_rad = np.pi / 180
arcmin_to_deg = 1 / 60
arcsec_to_arcmin = 1 / 60
arcsec_to_deg = arcsec_to_arcmin * arcmin_to_deg
arcsec_to_rad = arcsec_to_deg * deg_to_rad

print("=== Arcsecond to Radian Conversion ===")
print("1秒角は何ラジアンになるか？ / How many radians is 1 arcsecond?")
print()

# Step-by-step conversion
print("Step-by-step conversion:")
print("1. 1 arcsecond = 1/60 arcminutes = 1/60 分角")
print("2. 1 arcminute = 1/60 degrees = 1/60 度")
print("3. 1 degree = π/180 radians = π/180 ラジアン")
print()

print("Therefore / したがって:")
print("1 arcsecond = (1/60) × (1/60) × (π/180) radians")
print("1 秒角 = (1/60) × (1/60) × (π/180) ラジアン")
print()

# Calculate the exact value
exact_value = 1 / (60 * 60) * np.pi / 180
print(f"Exact calculation / 正確な計算:")
print(f"1 arcsecond = {exact_value:.10e} radians")
print(f"1 秒角 = {exact_value:.10e} ラジアン")
print()

# Compare with the given answer
given_answer = 4.8e-6
print(f"Given answer / 与えられた答え: {given_answer:.1e} radians")
print(f"Difference / 差: {abs(exact_value - given_answer):.2e} radians")
print()

# Create visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Left plot: Angular scale comparison
ax1.set_xlim(-2, 2)
ax1.set_ylim(-2, 2)
ax1.set_aspect('equal')

# Draw circles for different angular scales
angles_deg = [1, 1/60, 1/3600]  # 1 degree, 1 arcminute, 1 arcsecond
labels = ['1 degree (1度)', '1 arcminute (1分角)', '1 arcsecond (1秒角)']
colors = ['red', 'blue', 'green']

for i, (angle_deg, label, color) in enumerate(zip(angles_deg, labels, colors)):
    angle_rad = angle_deg * deg_to_rad
    
    # Draw arc
    if i == 0:  # 1 degree - visible arc
        arc = Arc((0, 0), 3, 3, angle=0, theta1=0, theta2=angle_deg, 
                 color=color, linewidth=2, label=label)
        ax1.add_patch(arc)
        
        # Add radius lines
        x_end = 1.5 * np.cos(angle_rad)
        y_end = 1.5 * np.sin(angle_rad)
        ax1.plot([0, 1.5], [0, 0], color=color, linewidth=1, alpha=0.7)
        ax1.plot([0, x_end], [0, y_end], color=color, linewidth=1, alpha=0.7)
    else:
        # For smaller angles, just show as text since they're too small to see
        ax1.text(-1.5, 1.5 - i*0.3, f'{label}: {angle_rad:.2e} rad', 
                color=color, fontsize=10, weight='bold')

ax1.set_title('Angular Scale Comparison\n角度スケール比較', fontsize=14, weight='bold')
ax1.grid(True, alpha=0.3)
ax1.legend()

# Right plot: Conversion factors visualization
ax2.set_xlim(0, 10)
ax2.set_ylim(0, 10)

# Create a flow chart showing conversion
steps = [
    "1 arcsecond\n1 秒角",
    "1/60 arcminute\n1/60 分角", 
    "1/3600 degree\n1/3600 度",
    "π/648000 radian\nπ/648000 ラジアン"
]

y_positions = [8, 6, 4, 2]
x_position = 2

for i, (step, y_pos) in enumerate(zip(steps, y_positions)):
    # Draw box
    box = plt.Rectangle((x_position-1, y_pos-0.5), 6, 1, 
                       facecolor='lightblue', edgecolor='black', alpha=0.7)
    ax2.add_patch(box)
    
    # Add text
    ax2.text(x_position+2, y_pos, step, ha='center', va='center', 
            fontsize=10, weight='bold')
    
    # Add arrow (except for last step)
    if i < len(steps) - 1:
        ax2.annotate('', xy=(x_position+2, y_positions[i+1]+0.5), 
                    xytext=(x_position+2, y_pos-0.5),
                    arrowprops=dict(arrowstyle='->', lw=2, color='red'))

# Add conversion factors on arrows
conversion_texts = ["÷ 60", "÷ 60", "× π/180"]
for i, text in enumerate(conversion_texts):
    ax2.text(x_position+3.5, (y_positions[i] + y_positions[i+1])/2, 
            text, fontsize=12, weight='bold', color='red',
            bbox=dict(boxstyle="round,pad=0.3", facecolor='white', alpha=0.8))

ax2.set_title('Conversion Steps\n変換ステップ', fontsize=14, weight='bold')
ax2.set_xticks([])
ax2.set_yticks([])
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['bottom'].set_visible(False)
ax2.spines['left'].set_visible(False)

plt.tight_layout()
plt.show()

# Additional calculations for understanding
print("\n=== Additional Information / 追加情報 ===")
print("Angular measurements hierarchy / 角度測定の階層:")
print("1 degree = 60 arcminutes = 3600 arcseconds")
print("1 度 = 60 分角 = 3600 秒角")
print()

print("In astronomy, arcseconds are used because:")
print("天文学で秒角が使われる理由:")
print("- Stars appear very small in the sky / 星は空で非常に小さく見える")
print("- Precise measurements needed / 精密な測定が必要")
print("- 1 arcsecond ≈ size of a coin 4 km away / 1秒角 ≈ 4km離れたコインの大きさ")
print()

print("Common astronomical angles / 一般的な天文角度:")
print(f"- Moon diameter: ~1800 arcseconds / 月の直径: ~1800秒角")
print(f"- Jupiter diameter: ~40 arcseconds / 木星の直径: ~40秒角")
print(f"- Typical star: <1 arcsecond / 一般的な星: <1秒角")
```
