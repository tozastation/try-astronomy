# 問1.6: プランク単位の理解と1秒角のラジアン変換
## 自然界の基本スケールと天体観測の角度単位

### 📖 問題文と学習目的

**問題1**: 1秒角はラジアンでなんなにか？

**問題2**: プランク時間 $t_p \equiv (G\hbar/c^5)^{1/2} = 1.4 \times 10^{-43}$ 秒などとも導けれる。

### 🎯 理論的背景と学習ポイント

**角度単位と物理基本単位の理解**
この問題は2つの重要な概念を扱います：天文学での角度測定とプランク単位系です。角度測定は天体観測の基礎であり、プランク単位は物理学の最も基本的なスケールを表します。

**重要な物理概念**
- **角度単位**: 度、分、秒角とラジアンの関係
- **プランク単位**: 基本物理定数から構成される自然単位系
- **基本物理定数**: 重力定数、プランク定数、光速
- **量子重力**: プランクスケールでの物理現象

**学習目標**
1. 角度単位の正確な変換方法の習得
2. プランク単位系の物理的意味の理解
3. 基本物理定数の組み合わせによる単位構成
4. 自然界の最小スケールの概念把握
